```python
#Extracting all the .pdf values

import os

values = os.listdir(r"C:\Users\Cash\Downloads")

values_pdf = list(filter(lambda x: x[-4:] == ".pdf", values))

values_pdf
```

```python
#Extracting all the values that have numbers
import os

values = os.listdir(r"C:\Users\Cash\Downloads")

def numeros(n):
    return any([x.isdigit() for x in n])

values_nums = list(filter(numeros, values))

values_nums
```

```python
#Extracting all the values that have no numbers
import os

values = os.listdir(r"C:\Users\Cash\Downloads")

def numeros(n):
    value = any([x.isdigit() for x in n])
    if value == True:
        return False
    else:
        return True

values_nonums = list(filter(numeros, values))

values_nonums
```
